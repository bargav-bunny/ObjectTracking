Checking `fps` for original video

```python
import cv2

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
else:
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"Frames per Second: {fps}")

    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    print(f"Frame Size: {frame_width} x {frame_height} pixels")

# Release the video capture object
cap.release()
```

```python
import os

def frames_to_video(frame_list, output_video_path, frame_size, fps=30):
    """
    Combines images into a video.

    Parameters:
    - frame_list: List of paths to the frames
    - output_video_path: Path where the output video will be saved
    - frame_size: Tuple of (width, height) for the output video
    - fps: Frames per second of the output video
    """
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, frame_size)

    for frame_path in frame_list:
        frame = cv2.imread(frame_path)
        if frame is None:
            print(f"Skipping invalid frame: {frame_path}")
            continue
        resized_frame = cv2.resize(frame, frame_size)
        out.write(resized_frame)

    out.release()
    print(f"Video saved to {output_video_path}")

frame_list = os.listdir(os.path.join(output_path, 'overlay'))
frame_list = [os.path.join(os.path.join(output_path, 'overlay'), x) for x in frame_list]

output_video_path = os.path.join(output_path, 'output_video.mp4')
frame_size = (1920, 1080)
fps = 30

frames_to_video(frame_list, output_video_path, frame_size, fps)
```
